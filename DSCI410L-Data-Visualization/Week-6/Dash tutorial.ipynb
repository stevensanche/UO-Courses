```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import dash
from dash import Dash, html, dash_table, dcc
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
```

```python
flu = pd.read_csv("./data/influenza_weekly.csv")
flu.head()
```

### Define an HTML layout

In this case, we create a single "Div", or section to include visual or interactive elements.

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    # Add a title to the division
    html.H1("Flu cases through time"),
    
    # Visual or interactive components to be added to the Div
])
```

### Add visual or interactive elements to your Div(s)

Here we add a scatterplot, as well as a Dropdown menu. These elements come from Dash Core Components (dcc). Each one has an associated ID, which is used as a reference for Inputs and Outputs.

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    # Add a title to the division
    html.H1("Flu cases through time"),

    # Add a dropdown menu to the Div. The options come from the data.
    dcc.Dropdown(
        id = "Region Select",
        options = np.unique(flu["FLUREGION"]),
        # Default value
        value = "North America",
        clearable=False
    ),

    # Add a barchart to the Div
    dcc.Graph(id = "barchart")
])
```

### Define components as either Inputs or Outputs

Pass componenets as either outputs (often graphs) or as inputs e.g. our dropdown menu.

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    # Add a title to the division
    html.H3("Flu cases through time"),
    
    # Add a dropdown menu to the Div. The options come from the data.
    dcc.Dropdown(
        id='Region Select',
        options = np.unique(flu["FLUREGION"]),
        # Default value
        value='North America',
        clearable=False
    ),

    # Add a barchart to the Div
    dcc.Graph(id = "barchart")
])

@app.callback(
    # Define the visual output
    Output("barchart", "figure"),
    # Define where user-input comes from
    Input("Region Select", "value")
)
```

### Define how Outputs incorporate Input

Create the associated output as a function, passing inputs as arguments.

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    # Add a title to the division
    html.H3("Flu cases through time"),
    
    # Add a dropdown menu to the Div. The options come from the data.
    dcc.Dropdown(
        id='Region Select',
        options = np.unique(flu["FLUREGION"]),
        # Default value
        value='North America',
        clearable=False
    ),

    # Add a barchart to the Div
    dcc.Graph(id = "barchart")
])

@app.callback(
    # Define the visual output
    Output("barchart", "figure"),
    # Define where user-input comes from
    Input("Region Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region):

    fig = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = "ALL_INF", width = 600, height = 500)

    return fig

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8051')
```

### Change the html layout to add more components

We can change the layout of our components by specifying rows and columns using Dash Bootstrap Components.

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            # Add a title to the division
            html.H1("Flu cases through time"),
    
            # Add a dropdown menu to the Div. The options come from the data.
            dcc.Dropdown(
                id='Region Select',
                options = np.unique(flu["FLUREGION"]),
                # Default value
                value='North America',
                clearable=False
            ),

            # Add a barchart to the Div
            dcc.Graph(id = "barchart")
            
        ], width = 5),
        dbc.Col([dcc.Graph(id = "scatterplot")
              ], width = 5)
    ])
])

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = "ALL_INF", width = 500, height = 500)
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = "ALL_INF", width = 500, height = 500)

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8051')
```

### Adding multiple interactive components

Just as we can add multiple graphs to the dashboard, we can add multiple interactive components to provide further user-control. Here we'll adjust the y variable by giving options of type of flu to choose from.

```python
# Create options for a new dropdown menu
# If we want to have different labels for values in the data, we need key-value pairs.
flu_types = {"INF_A": "Influenza A", "INF_B": "Influenza B", "AH1": "Swine Flu", "ALL_INF": "All types"}
flu_types
```

Note that inputs are passed as arguments to your output functions in order of the Inputs in your Callback.

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container([
    dbc.Row([# Add a title to the division
            html.H1("Flu cases through time"),
    
            # Add a dropdown menu to the Div. The options come from the data.
            dcc.Dropdown(
                id='Region Select',
                options = np.unique(flu["FLUREGION"]),
                # Default value
                value='North America',
                clearable=False
            ),
            html.H4("Flu type"),
                dcc.Dropdown(
                id='Flu Select',
                options = flu_types,
                # Default value
                value='ALL_INF',
                clearable=False)]),
    dbc.Row([
        dbc.Col([
            # Add a barchart to the Div
            dcc.Graph(id = "barchart")
            
        ], width = 5),
        dbc.Col([dcc.Graph(id = "scatterplot")
              ], width = 5)
    ])
])

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value"),
    Input("Flu Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region, flu_type):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = flu_type, width = 500, height = 500)
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = flu_type, width = 500, height = 500)

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8051')
```
