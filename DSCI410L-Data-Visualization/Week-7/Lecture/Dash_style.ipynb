```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import dash
from dash import Dash, html, dash_table, dcc
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
```

```python
flu = pd.read_csv("./data/influenza_weekly.csv")
flu_types = {"INF_A": "Influenza A", "INF_B": "Influenza B", "AH1": "Swine Flu", "ALL_INF": "All types"}
```

Here is our dashboard as we left it last time. 

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container([
    dbc.Row([# Add a title to the division
            html.H1("Flu cases through time"),
    
            # Add a dropdown menu to the Div. The options come from the data.
            dcc.Dropdown(
                id='Region Select',
                options = np.unique(flu["FLUREGION"]),
                # Default value
                value='North America',
                clearable=False
            ),
            html.H4("Flu type"),
                dcc.Dropdown(
                id='Flu Select',
                options = flu_types,
                # Default value
                value='ALL_INF',
                clearable=False)]),
    dbc.Row([
        dbc.Col([
            # Add a barchart to the Div
            dcc.Graph(id = "barchart")
            
        ], width = 5),
        dbc.Col([dcc.Graph(id = "scatterplot")
              ], width = 5)
    ])
])

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value"),
    Input("Flu Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region, flu_type):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = flu_type, width = 500, height = 500)
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = flu_type, width = 500, height = 500)

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8060')
```

It's perfectly fine, though maybe a little visually dull. Also, it doesn't translate incredibly well to a full browser webpage. We'll now explore how to adjust the styling of our dashboard using CSS.

## Pre-set Themes

The first and simplest way to changet the style of your app is with a pre-set theme. Dash Bootstrap has a number of pre-set themes you can apply to your dashboards easily ([found here](https://dash-bootstrap-components.opensource.faculty.ai/docs/themes/#available-themes)). Essentially, these change the background color of certain elements and/or font styles.

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.SKETCHY])

app.layout = dbc.Container([
    dbc.Row([# Add a title to the division
            html.H1("Flu cases through time"),
    
            # Add a dropdown menu to the Div. The options come from the data.
            dcc.Dropdown(
                id='Region Select',
                options = np.unique(flu["FLUREGION"]),
                # Default value
                value='North America',
                clearable=False
            ),
            html.H4("Flu type"),
                dcc.Dropdown(
                id='Flu Select',
                options = flu_types,
                # Default value
                value='ALL_INF',
                clearable=False)]),
    dbc.Row([
        dbc.Col([
            # Add a barchart to the Div
            dcc.Graph(id = "barchart")
            
        ], width = 5),
        dbc.Col([dcc.Graph(id = "scatterplot")
              ], width = 5)
    ])
])

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value"),
    Input("Flu Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region, flu_type):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = flu_type, width = 500, height = 500)
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = flu_type, width = 500, height = 500)

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8060')
```

## Inline CSS

The pre-set themes can be a decent starting point, but you will unavoidably need to tweak CSS elements yourself.

A fast way of adjusting the appearance of your apps is with inline CSS. This is where we make one-off modifications to the style of a specific html element using the `style` argument. There are [many properties of CSS](https://htmlcheatsheet.com/css/). 

Here we add background colors to each of our Divs, change the font and color of our HTML headers, and style the Container to be "fluid", which means it will fill the entire webpage. I also change the colors of the graphs and fonts to look better on the background.

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container([
    dbc.Row([# Add a title to the division
            html.H1("Flu cases through time", style = {"fontFamily": "Helvetica", "color":"White"}),
    
            # Add a dropdown menu to the Div. The options come from the data.
            dcc.Dropdown(
                id = 'Region Select',
                options = np.unique(flu["FLUREGION"]),
                # Default value
                value = 'North America',
                clearable = False
            ),
            html.H4("Flu type", style = {"fontFamily": "Helvetica", "color":"White"}),
                dcc.Dropdown(
                id='Flu Select',
                options = flu_types,
                # Default value
                value = 'ALL_INF',
                clearable = False)]),
    dbc.Row([
        dbc.Col([
            # Add a barchart to the Div
            dcc.Graph(id = "barchart")
            
        ], width = 6, 
           style={'backgroundColor': '#074050'}),
        
        dbc.Col([dcc.Graph(id = "scatterplot")
              ], width = 6, 
                style = {'backgroundColor': '#074050'})
    ])
    
], fluid = True, style = {'backgroundColor': '#074050'})

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value"),
    Input("Flu Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region, flu_type):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = flu_type, width = 450, height = 500)

    fig1.update_layout(
        paper_bgcolor = "#074050",
        font=dict(color = "white")
    )
    
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = flu_type, width = 450, height = 500)

    fig2.update_layout(
        paper_bgcolor = "#074050",
        font=dict(color = "white")
    )

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8060')
```

### Adjusting layout and spacing

I don't personally like the long dropdown bars, nor do I like the spacing between the two graphs. 

Here we change the layout such that our drop-down bars are spaced in three rows within a single column to the left of the figures. We add padding by applying a specific class with `className`  to increase the vertical distance between each row in the first column (see "gutters" in the link below). The maximum width of a row is 12, so we change the width of each of our three columns to 4.

For options for full layout control, see [this page](https://dash-bootstrap-components.opensource.faculty.ai/docs/components/layout/).

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.CERULEAN])

app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            dbc.Row([html.H1("Flu cases through time", style = {"fontFamily": "Helvetica", "color":"White"})], className = "mb-3"),
                     
            dbc.Row([html.H5("Geographic Region", style = {"fontFamily": "Helvetica", "color":"White"}),
                     dcc.Dropdown(
                    id = 'Region Select',
                    options = np.unique(flu["FLUREGION"]),
                    value = 'North America',
                    clearable = False)
                    ], className = "mb-3"),
                     
            dbc.Row([html.H5("Flu Type", style = {"fontFamily": "Helvetica", "color":"White"}),
                    dcc.Dropdown(
                    id = 'Flu Select',
                    options = flu_types,
                    value = 'ALL_INF',
                    clearable = False)])
                ], width = 3),
        dbc.Col([
            # Add a barchart to the Div
            dcc.Graph(id = "barchart")
            
        ], width = 4, 
           style = {'backgroundColor': '#074050'}),
        
        dbc.Col([dcc.Graph(id = "scatterplot")
              ], width = 4, 
                style = {'backgroundColor': '#074050'})
    ])
    
], fluid = True, style = {'backgroundColor': '#074050'})

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value"),
    Input("Flu Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region, flu_type):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = flu_type, width = 450, height = 500)

    fig1.update_layout(
        paper_bgcolor = "#074050",
        font=dict(color = "white"),
        title = "Flu cases by year"
    )
    
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = flu_type, width = 450, height = 500)

    fig2.update_layout(
        paper_bgcolor = "#074050",
        font=dict(color = "white"),
        yaxis_title = "",
        title = "Flu cases by week"
    )

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8060')
```

Note however that this still looks a little funny when opened in a web browser. This is best fixed using a style sheet. 

## Assets and stylesheets

All stylings that we've defined inline in the above examples can instead be coded as classes in a CSS stylesheet. These classes can then be applied to each HTML element if the stylesheet, referenced by default as "styles.css", is present in an associated "assets" folder. This is the suggested approach when you intend to repeat a lot of the same styles in your HTML elements. 

I've made many changes to the app, most of which are contained within the .css style sheet. Note that we've remove nearly all inline CSS, replacing it with references to classes in the style sheet instead.  

Also note that I've made a number of changes to the figure code. The most important is Plotly's "autosize" argument in `update_layout()`. This is specifically targeted at web apps, such that graphs will automatically resize themselves to fill their Divs. This ensures that your layouts and figure sizes stay consistent across different screen sizes. I've also changed the colors of the background and markers of the figures.

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            dbc.Row([html.H1("Flu cases through time", className = "label-font"), 
                     html.P("These data show the total number of flu cases (Inf. A, B and Swine Flu) in different regions of the world.",
                            className = "paragraph-text")
                    ], className="mb-3"),
                     
            dbc.Row([html.H5("Geographic Region", className = "label-font"),
                     dcc.Dropdown(
                    id = 'Region Select',
                    options = np.unique(flu["FLUREGION"]),
                    value = 'North America',
                    clearable = False)
                    ], className = "mb-3"),
                     
            dbc.Row([html.H5("Flu Type", className="label-font"),
                    dcc.Dropdown(
                    id = 'Flu Select',
                    options = flu_types,
                    value='ALL_INF',
                    clearable=False)])
                    ], width = 2, className = "dropdown-background"),
        dbc.Col([
            dcc.Graph(id = "barchart", className = "graph-shadow")
        ], width = 4, className = "graph-background"),
        
        dbc.Col([
            dcc.Graph(id = "scatterplot", className = "graph-shadow")
        ], width = 4, className = "graph-background")
    ], className = "first-row-height")
    
], fluid = True, className = "app-background")

@app.callback(
    # Define the visual output
    [Output("barchart", "figure"),
    Output("scatterplot", "figure")],
    # Define where user-input comes from
    Input("Region Select", "value"),
    Input("Flu Select", "value")
)

# Arguments passed to functions are in order of Inputs 
def flu_plot(region, flu_type):

    fig1 = px.bar(flu[flu["FLUREGION"] == region], x = "Year", y = flu_type)

    fig1.update_layout(
        paper_bgcolor = "#084d61",
        plot_bgcolor = "#084d61",
        font = dict(color="white"),
        yaxis_title = f"{flu_types[flu_type]} Cases",
        title = "Flu cases by year",
        autosize = True
    )
    
    fig1.data[0].marker.color = "#d3f2a3"
    fig1.data[0].marker.line.color = "black"
    
    fig2 = px.scatter(flu[flu["FLUREGION"] == region], x = "Week", y = flu_type)

    fig2.update_layout(
        paper_bgcolor = "#084d61",
        plot_bgcolor = "#084d61",
        font = dict(color = "white"),
        yaxis_title="",
        title = "Flu cases by week",
        autosize = True
    )
    
    fig2.data[0].marker.color = "#d3f2a3"

    return fig1, fig2

# Run the app
if __name__ == '__main__':
    app.run(debug=True, port = '8060')
```
