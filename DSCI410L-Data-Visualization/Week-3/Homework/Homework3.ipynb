### Figure 1:

```python
data = {'Project Type': ['Transmission & Distribution', 'Treatment', 'Storage', 'Source', 'Other'],
        'Dollars': [200.80, 75.10, 36.90, 6.00, 2.30]
       }
                         
figure1 = pd.DataFrame(data)
figure1.sort_values(by = 'Dollars', ascending = False, inplace = True)
figure1
```

```python
fig, ax = plt.subplots(figsize = (12,8))
sns.barplot(data = figure1, x = 'Project Type', y = 'Dollars', color = 'green')

plt.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.xticks(fontsize = 10);
plt.yticks([]);

plt.ylabel('')
plt.xlabel('')

trans = figure1.iloc[0]
treat = figure1.iloc[1]
storage = figure1.iloc[2]
source = figure1.iloc[3]
other = figure1.iloc[4]

ax.text(0, trans['Dollars'] + 1, 
        '$' + trans['Dollars'].astype(str), ha = 'center')

ax.text(1, treat['Dollars'] + 1, 
        '$' + treat['Dollars'].astype(str), ha = 'center')

ax.text(2, storage['Dollars'] + 1,
        '$' + storage['Dollars'].astype(str), ha = 'center')

ax.text(3, source['Dollars'] + 1,
        '$' + source['Dollars'].astype(str), ha = 'center')

ax.text(4, other['Dollars'] + 1, 
        '$' + other['Dollars'].astype(str), ha = 'center')

plt.title('Infrastructure needs of water systems in 2007 (in billions USD)',
         fontsize = 14);
```

#### Explanation to Figure 1: 

I think that this figure is a lot better than the original becuase it gets rid of a lot of the uneccessary ink, like the fading bars and the grid. I chose green because we are talking about money, so I feel like it was the perfect color. Also, since we already have the exactly amount of money needed for each project type, then there is no reason for a y label to be there, and we can just eliminate that and keep the values on each bar. We also kept the x labels since they differentiate each project type, and just added a title, which covers what the graph is about, what year, and the dollar scale.

```python
data4 = pd.read_excel('linegraph_data.xlsx')

data4.head()
```

```python
data4['medium'].unique()
```

```python
website = data4[data4['medium'] == 'website'].sort_values(by = 'date')
digital_ads = data4[data4['medium'] == 'digital ads'].sort_values(by = 'date')
email = data4[data4['medium'] == 'email'].sort_values(by = 'date')
events = data4[data4['medium'] == 'events'].sort_values(by = 'date')
list_acquisition = data4[data4['medium'] == 'list acquisition'].sort_values(by = 'date')
mobile = data4[data4['medium'] == 'mobile'].sort_values(by = 'date')
social = data4[data4['medium'] == 'social'].sort_values(by = 'date')
```

```python
fig, ax = plt.subplots(7, 1, figsize = (10, 20))

fig.suptitle('Platform Revenue Over Time in 2025', y = 0.93, fontsize = 14)
fig.supxlabel('Date', y = 0.07, fontsize = 14)
fig.supylabel('Revenue (in thousands USD)', x = 0.06, fontsize = 14)

ax[0].plot(website['date'], website['revenue_thousands'], marker = 'o')
ax[0].set_xticks([])
ax[0].set_title('Website')


ax[1].plot(digital_ads['date'], digital_ads['revenue_thousands'], marker = 'o')
ax[1].set_xticks([])
ax[1].set_title('Digital Ads')

ax[2].plot(email['date'], email['revenue_thousands'], marker = 'o')
ax[2].set_xticks([])
ax[2].set_title('Email')

ax[3].plot(events['date'], events['revenue_thousands'], marker = 'o')
ax[3].set_xticks([])
ax[3].set_title('Events')

ax[4].plot(list_acquisition['date'], list_acquisition['revenue_thousands'], marker = 'o')
ax[4].set_xticks([])
ax[4].set_title('List Acquisition')

ax[5].plot(mobile['date'], mobile['revenue_thousands'], marker = 'o')
ax[5].set_xticks([])
ax[5].set_title('Mobile')

ax[6].plot(social['date'], social['revenue_thousands'], marker = 'o')
ax[6].set_title('Social')

plt.savefig('GraphLowRes.png', dpi = 70)
plt.show()
```

### Figure 5:

```python
data5 = [
    {'Source': 'Fox News', 'Jan 2020': 111565, 'Jan 2024': 85024},
         {'Source': 'CNN', 'Jan 2020': 141008, 'Jan 2024': 112754},
         {'Source': 'The New York Times', 'Jan 2020': 103670, 'Jan 2024': 51235},
         {'Source': 'The Washington Post', 'Jan 2020': 87890, 'Jan 2024': 49624}
        ]

figure5 = pd.DataFrame(data5)

figure5
```

```python
fig, ax = plt.subplots(figsize = (14, 7))

bar_width = 0.4

x_positions = np.arange(len(figure5['Source']))

ax.bar(x_positions - bar_width/2, figure5['Jan 2020'], bar_width, label = 'Jan 2020', color = 'steelblue')
ax.bar(x_positions + bar_width/2, figure5['Jan 2024'], bar_width, label = 'Jan 2024', color = 'tomato')

ax.set_xticks(x_positions);
ax.set_xticklabels(figure5['Source'], fontsize = 12);

ax.set_xlabel('Sources', fontsize = 15)
ax.set_ylabel('Visitors', fontsize = 15)
ax.set_title('Unique Visitors for Different News Outlets from 2020 - 2024', fontsize = 18)

ax.legend(frameon = False, fontsize = 12)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

#### Explanation to Figure 5:

For this one, I think that having a side by side bar plot with the visual differences was the best way to plot this. We can easily see how each news outlet had a decline in viewership, and we can see how it's represented by the numbers of visitors on the side. I also added a legend to easily differentiate each year, and I think that this is very minimal and well ratio'ed to the amount of data we are working with.
