We'll use this "crop recommendation" dataset, which lists a number of crops recommended for given nutrient and environmnetal conditions, derived from agricultural data in India.

```python
crops= pd.read_csv("./data/Crop_recommendation.csv")
crops["water_need"] = np.where(crops["rainfall"] < crops["rainfall"].median(), "low", "high")
crops["soil_acidity"] = np.where(crops["ph"] < 7, "acidic", "basic")
crops.head()
```

#### Part 1: Minimize the Data to Ink Ratio

Using a default matplotlib or seaborn plot as a point of reference, dramatically increase the Data to Ink ratio for your chosen plot. By this I *do not* mean make a bad plot; you should still strive to make an effective plot, but one that uses quite a bit more ink than your typical seaborn default to improve some combination of clarity and beauty. 

A  (non-exhaustive) list of things you can edit for this and following questions:

- Gridlines
- Tick properties
- Background color
- Axis, grid, and label color
- Size of encodings (points, bars, boxes, etc...)
- Legends
- Axis spines on all sides
- Titles, subtitles, captions
- Markers, outlines, fills
- Faceting

```python
### Your plot
fig, ax = plt.subplots(figsize=(13, 10))

ax = plt.gca()
ax.set_facecolor("black")
plt.grid(False)


sns.scatterplot(data = crops, x = 'P', y = 'label', hue = 'soil_acidity', 
                palette = ['#ff9f9b', '#8de5a1'], edgecolors = 'grey', s = 50,
               alpha = 0.8, marker = '.')


mean = crops.groupby('label')['P'].mean().reset_index()
ax.scatter(mean['P'], mean['label'], label = 'p-mean', color = 'red', s = 30, marker = 'o', edgecolor = 'black')


plt.legend(bbox_to_anchor = (1, 1), fontsize = 11, markerscale = 1.4)
plt.grid(True, alpha = 0.4)
ax.xaxis.grid(True, alpha = 0.2) 


plt.title('Phosphorus of each crop', fontsize = 14, fontweight = 'bold', fontfamily = 'serif')
plt.xlabel('Phosphorus', fontweight = 'bold', fontfamily = 'serif', fontsize = 12)
plt.ylabel('Crops', fontweight = 'bold', fontfamily = 'serif', fontsize = 12)
```

#### Part 2: Optimize the Data to Ink Ratio

Now find the right ratio that you think perfectly balances data and ink. Display the data clearly, and use beautifying ink sparingly and purposefully.

```python
### Your plot

fig, ax = plt.subplots(figsize=(13, 10))



ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

sns.scatterplot(data = crops, x = 'P', y = 'label', hue = 'soil_acidity', 
                palette = ['#ff9f9b', '#8de5a1'], edgecolors = 'grey', s = 50,
               alpha = 0.8, marker = '.')


mean = crops.groupby('label')['P'].mean().reset_index()
ax.scatter(mean['P'], mean['label'], label = 'p-mean', color = 'black', s = 20, marker = 'o', edgecolor = 'grey')


plt.legend(loc = 'best')
plt.grid(True, alpha = 0.3)
ax.xaxis.grid(True, alpha = 0.3) 


plt.title('Phosphorus of each crop', fontsize = 14, fontweight = 'bold', fontfamily = 'serif')
plt.xlabel('Phosphorus', fontweight = 'bold', fontfamily = 'serif', fontsize = 12)
plt.ylabel('Crops', fontweight = 'bold', fontfamily = 'serif', fontsize = 12)
```

#### Part 3: Maximizer the Data to Ink Ratio

Lastly, maximize the ratio i.e. show as much data as you can with as little ink as possible. Your goal here is minimalism without sacrificing *too much* information or clarify, though some sacrifices may be warranted. 

Remember, this is a balance of both data and ink. You can remove virtually all visual elements of a plot but if you've also removed too much data, then you've decreased your ratio by decreasing data. **The group that achieves the largest data to ink ratio will receieve 20% extra credit on this lab**.

```python
### Your plot
fig, ax = plt.subplots(figsize=(8, 10))



ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

sns.scatterplot(data = crops, x = 'P', y = 'label', hue = 'soil_acidity', 
                palette = ['grey', 'black'], edgecolors = 'none', s = 25,
               alpha = 0.8, marker = '.')


mean = crops.groupby('label')['P'].mean().reset_index()
ax.scatter(mean['P'], mean['label'], label = 'Phosphorus mean', color = 'tomato', s = 15, marker = 'o', edgecolor = 'none')


plt.legend(bbox_to_anchor = (1.4, 1), markerscale = 1.8, fontsize = 9)
plt.grid(False)
ax.xaxis.grid(False)

ax.tick_params(axis='both', which='both', length=0)
ax.tick_params(axis='x', labelsize = 8, labelfontfamily = 'serif')
ax.tick_params(axis='y', labelsize = 8, labelfontfamily = 'serif')

plt.title('Phosphorus of each crop', fontsize = 14, fontweight = 'light', fontfamily = 'serif')
plt.xlabel('Phosphorus', fontweight = 'light', fontfamily = 'serif', fontsize = 12)
plt.ylabel('Crops', fontweight = 'light', fontfamily = 'serif', fontsize = 12)
```
