```python
import pandas as pd
import numpy as np
import plotly.express as px
import dash
from dash import Dash, html, dash_table, dcc
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
```

# Just run this app

```python
nba = pd.read_csv("three_point_revolution.csv")
nba["Year"] = nba["Season"].str.split("-", expand = True)[0].astype(int)
nba = nba.sort_values("Year")
nba.head()
```

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            dcc.RangeSlider(
                id = "y-slider",
                min = 0,
                max = 50,
                step = 1,
                marks = {i: str(i) for i in range(0, 50, 5)},
                value = [0, 45],
                vertical = True,
                updatemode = 'drag'
            )
        ], width = 1),
        dbc.Col([
            dcc.Graph(id = "lineplot")
        ], width = 4)
    ]),
        dbc.Row([
        dbc.Col([
        ], width = 2),
        dbc.Col([
            dcc.RangeSlider(
                id = "x-slider",
                min = 1950,
                max = 2025,
                step = 1,
                marks = {i: str(i) for i in range(1950, 2025, 5)},
                value = [1970, 2025],
                updatemode = 'drag'
            )
        ], width = 8)
    ])
])

@app.callback(
    Output("lineplot", "figure"),
    Input("y-slider", "value"),
    Input("x-slider", "value")
)

def update_lineplot(y_range, x_range):

    y_min, y_max = y_range
    x_min, x_max = x_range
    
    fig = px.line(nba[nba["Year"] > 1970], x = "Year", y = "3PA", width = 700, height = 500)

    fig.update_layout(
        yaxis_title = "3-Point Attempts",
        yaxis = dict(range = [y_min, y_max]),
        xaxis = dict(range = [x_min, x_max])
    )

    return fig

if __name__ == '__main__':
    app.run(debug=True)
```
