## Color Palettes

```python
inj = pd.read_csv("./data/collegiate_athlete_injury.csv")
inj.head()
```

When plotting data with seaborn, it will reach for the default color palette.

```python
sns.color_palette()
```

While qualitative color palettes are more forgivign when choosing representative colors, the right color palettes are essential other certain types of figures. Heatmaps are one such example. As opposed to palettes, they utilize a continue range of color called a color map or `cmap`.

```python
sns.heatmap(inj.drop(columns = ["Athlete_ID", "Gender", "Position"]).corr(), vmin=-1, vmax=1)
```

Just like palettes, there are pre-defined color maps and you can make your own.

```python
sns.color_palette("viridis", as_cmap=True)
```

**What colors would you choose?**

```python
better_heatmap_palette = sns.color_palette("blend:#0d16ba,#ba190d", as_cmap=True)
better_heatmap_palette
```

```python
sns.heatmap(inj.drop(columns = ["Athlete_ID", "Gender", "Position"]).corr(), vmin=-1, vmax=1, cmap = better_heatmap_palette)
```

`color_palette` will by default create a sequential palette between two colors. Two create a diverging palette, use `diverging_palette()`. One thing to be mindful of is that `diverging_palette` takes colors as "Hue Saturation and Lightness", or "HSL", values. The first two arguments are the end-point colors as angular values (from a color wheel).

```python
better_heatmap_palette = sns.diverging_palette(200, 20, as_cmap=True)
better_heatmap_palette
```

```python
sns.heatmap(inj.drop(columns = ["Athlete_ID", "Gender", "Position"]).corr(), vmin=-1, vmax=1, cmap = better_heatmap_palette)
```
