### Figure 1: Challenge 3 | Red Category

```python
challenge3 = pd.read_csv('challenge3.csv')

challenge3 = challenge3.sort_values(by = 'Date', ascending = False)
challenge3.head()

challenge3['Date String'] = challenge3['Date'].astype(str) 
# converting to string since we are wanting a categorial y axis, not a continous one. With
# having it be integers, it will automatically jump by certain numbers since it is continous.
# when we switch it to a string, each date will now have its own category on the y axis


challenge3.head()
```

```python
# p = so.Plot(
#     data = challenge3, x = 'Land', y = 'Date String')

# p = p.label(x = '', y = '', title = 'ACRES OF LAND OWNED BY NEGROES IN GEORGIA')

# p = p.add(
#     so.Bar(color = 'red', width = 0.7))
```

```python
fig, ax = plt.subplots(figsize = (8, 10))

ax.barh(challenge3['Date String'], challenge3['Land'], color = 'tomato', height = 0.8)

ax.set_title('ACRES OF LAND OWNED BY NEGROES IN GEORGIA.',
             fontstyle = 'normal',
             fontfamily = 'serif',
             fontweight = 'normal',
            loc = 'center', 
             color = 'black'
            )

ax.set_xticks([])
plt.yticks(fontstyle = 'normal', fontfamily = 'serif', fontweight = 'light')


plt.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

top, bottom = challenge3.iloc[-1], challenge3.iloc[0]

ax.text(100000, top['Date String'], "338,769", 
        fontfamily = 'serif', fontweight = 'bold', fontsize = 10,
       va = 'center', ha = 'left');

ax.text(400000, bottom['Date String'], "1,062,223",
        fontfamily = 'serif', fontweight = 'bold', fontsize = 10,
       va = 'center', ha = 'left');
```

### Figure 3: Challenge 7 | Green Category

```python
challenge7 = pd.read_csv('challenge7.csv')
challenge7
```

```python
# the graph from the challenge is counting the column titles from the table as well, so 
# we have to include it in the data

new_row = pd.DataFrame({'Place': ['Romaine'],'Number': [73.3216]})

challenge7.columns = ["Place", "Number"]

challenge7 = pd.concat([new_row, challenge7])

challenge7 = challenge7.sort_values(by = 'Number', ascending = True)

redchallenge7 = challenge7[challenge7['Place'] == 'Negroes, U.S.A.']
```

```python
fig,ax = plt.subplots(figsize = (7, 8))

ax.barh(challenge7['Place'], challenge7['Number'], color = '#34934d', height = 0.8)
ax.barh(redchallenge7['Place'], redchallenge7['Number'], color = 'red', height = 0.8)


plt.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.suptitle('Illiteracy of the American Negroes compared with that of other nations.',
             fontweight = 'light', fontfamily = 'serif', color = 'black', fontsize = 11)

ax.set_title("Proportion d' illettres parmi les Negres Americains comparee a celle des autres nations.",
             fontweight = 'light', fontfamily = 'serif', color = 'black', fontsize = 8)

fig.text(.5, .9, 'Done by Atlanta University.', color = 'black', ha = 'center', fontweight = 'light', fontsize = 8)

plt.yticks(fontstyle = 'normal', fontfamily = 'serif', fontweight = 'light', fontsize = 7)
plt.xticks([])

plt.tight_layout()
```
