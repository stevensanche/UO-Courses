```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import shapely as shp
```

## Geometries

```python
p1 = shp.Point([0,0])
p1
```

```python
p2 = shp.Point([1,0])
p3 = shp.Point([1,1])
p4 = shp.Point([0,1])
```

```python
shp.LineString([p1,p2])
```

```python
shp.LineString([p1,p2,p3])
```

```python
shp.LineString([p1,p3,p2])
```

```python
shp.LineString([p1,p2,p3,p4])
```

```python
shp.Polygon((p1,p2,p3))
```

```python
square = shp.Polygon((p1,p2,p3,p4))
square
```

```python
square.area
```

```python
square.boundary
```

```python
square.centroid.xy
```

```python
shp.MultiPoint([p1,p2,p3,p4, square.centroid.xy])
```
