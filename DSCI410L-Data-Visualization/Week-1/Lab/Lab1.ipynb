```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# plt.style.use('default') '... .txt')
```

```python
df = pd.read_csv('car_insurance_premium_dataset.csv')
df
```

```python
no_accidents = df[df['Previous Accidents'] == 0]
one_accident = df[df['Previous Accidents'] == 1]
over_one = df[df['Previous Accidents'] > 1]
```

```python
zero = no_accidents.groupby('Driver Age')['Insurance Premium ($)'].mean().reset_index()
one = one_accident.groupby('Driver Age')['Insurance Premium ($)'].mean().reset_index()
multiple = over_one.groupby('Driver Age')['Insurance Premium ($)'].mean().reset_index()
```

```python
year = df.groupby('Driver Age')['Insurance Premium ($)'].mean().reset_index()

# plt.style.use('better_stylesheet.txt')
# plt.plot(year['Driver Age'], year['Insurance Premium ($)'])
```

```python
fig, ax = plt.subplots(3, 1, figsize = (7, 10))

plt.style.use('style.txt')

ax[0].plot(zero['Driver Age'], zero['Insurance Premium ($)'])
ax[0].set_title('Average cost of insurance per age (0 accidents)')

ax[1].plot(one['Driver Age'], one['Insurance Premium ($)'])
ax[1].set_title('Average cost of insurance per age (1 accident)')


ax[2].plot(multiple['Driver Age'], multiple['Insurance Premium ($)'])
ax[2].set_title('Average cost of insurance per age (multiple accidents)')

fig.supxlabel('Age')
fig.supylabel('Insurance price (USD)')
fig.tight_layout()
```
