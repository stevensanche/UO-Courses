```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gd = pd.read_csv("./data/global_development.csv")
gd.head()
```

```python
gd_2013 = gd[gd["Year"] == 2013]
```
