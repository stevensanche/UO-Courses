```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np

X, y = make_regression(n_samples = 1000, n_features = 20, noise = 10, random_state = 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                                                    random_state = 0)

print('Mean of features', np.mean(X, axis = 0))
print('STD of features', np.std(X, axis = 0))

mlp = MLPRegressor(hidden_layer_sizes = (64, 32), activation = 'relu',
                   learning_rate_init = 0.01, solver = 'sgd')

mlp.fit(X_train, y_train)

plt.plot(mlp.loss_curve_)
plt.xlabel('Epoch')
plt.ylabel('Training loss')
plt.title('MLPRegressor training loss')
plt.show()

print('R^2:', mlp.score(X_test, y_test))
```
