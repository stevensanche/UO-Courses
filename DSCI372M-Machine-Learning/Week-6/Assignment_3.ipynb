```python
# here, using project2_vision for guideline 

conv = nn.Conv2d(in_channels = 1, out_channels = 1, kernel_size = 3, stride = 2, padding = 1, bias = False)

# A torch.Tensor is a multi-dimensional matrix containing elements of a single data type. 
# Please see torch.dtype for more details about dtype support.
# https://docs.pytorch.org/docs/stable/tensors.html
Filter = torch.tensor([[
    [
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
    ]
]]).float()

image = torch.tensor([[
    [
    [3, 4, 2, 5, 6, 2, 3],
    [1, 7, 2, 3, 3, 6, 7],
    [2, 5, 3, 6, 5, 5, 6],
    [2, 4, 4, 7, 6, 7, 5],
    [3, 6, 3, 6, 7, 8, 6],
    [4, 5, 5, 5, 7, 6, 7],
    [6, 4, 3, 6, 6, 7, 7]
    ]
]]).float()

conv.weight.data = Filter
output = conv(image)

print(output[0][0])
```
