   y^{hat} = w_0 + w_1x_1 + w_2x_2

#### A)

1: $ (1, 1, -0.8) $

2: $ -1, -2, 0.1) $

3: $ (2, -1, -5.0) $

$y^{hat(1)} = 0 + 0(1) + 0(1) = 0 $

$y^{hat(2)} = 0 + 0(-1) + 0(-2) = 0 $

$y^{hat(3)} = 0 + 0(2) + 0(-1) = 0 $

errors: ($ y^{hat} - y $)
- $y^{hat(1)}: 0 - (-0.8) = 0.8 $
- $y^{hat(2)}: 0 - 0.1 = -0.1 $
- $y^{hat(3)}: 0 - (-5.0) = 5.0 $

```python
import numpy as np
import pandas as pd
```

```python
x = np.array([
    [1, 1],
    [-1, -2],
    [2, -1]
])

y = np.array([-0.8, 0.1, -5.0])

w0, w1, w2 = 0.0, 0.0, 0.0
alpha = 0.05
n = len(y)

for i in range(1, 5): #running 4 iterations
    y_hat = w0 + (w1 * x[:, 0]) + (w2 * x[:, 1])
    error = y_hat - y
    mean_squared_error = np.mean(error ** 2)
    gradient1 = (2/n) * np.sum(error)
    gradient2 = (2/n) * np.sum(error * x[:, 0])
    gradient3 = (2/n) * np.sum(error * x[:, 1])
    w0 = w0 - alpha * gradient1
    w1 = w1 - alpha * gradient2
    w2 = w2 - alpha * gradient3
    print(w0, w1, w2, mean_squared_error)
```

```python
X = np.array([
    [1, 1, 1],
    [1, -1, -2],
    [1, 2, -1]
])

xtx = X.T @ X
xtx_inverse = np.linalg.inv(xtx)

xty = X.T @ y

w = xtx_inverse @ xty

y_pred = X @ w

errors = y_pred - y 

total_error = np.sum(errors ** 2)

w, total_error
total_error
```

```python
xx = np.array([
    [1, 1],
    [2, -1],
    [-3, -1],
    [-3, 1]
])

yy = np.array([1, 0, 1, 1])

w = np.array([0, 0])

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def average_log_likelihood(w):
    log_likelihood = 0
    for a, b in zip(xx, yy):
        z = w @ a
        p = sigmoid(z)
        log_likelihood += b * np.log(p) + (1 - b) * np.log(1 - p)
    return log_likelihood / len(yy)


def gradient_ascent(alpha, n = 4):
    guesses = [np.array([0, 0])]
    current_guess = guesses[0]
    while len(guesses) < n + 1:
        gradient = np.array([0.0, 0.0])
        for a, b in zip(xx, yy):
            z = current_guess @ a
            p = sigmoid(z)
            gradient += (b - p) * a
        gradient = gradient / len(yy)
        current_guess = current_guess + alpha * gradient
        guesses.append(current_guess)
    return np.array(guesses)
```

```python
iteration1 = gradient_ascent(0.01, n = 4)
iteration2 = gradient_ascent(0.2, n = 4)
w1 = iteration1[-1] #last element
w2 = iteration2[-1]
ll1 = average_log_likelihood(w1)
ll2 = average_log_likelihood(w2)
```
