```python
import pandas as pd
import sqlalchemy
```

#### SQLite to create a DB

The next cell creates an association with an SQLite database named 'tech_ceo_info.db' in the current directory. When I executed the command, 'tech_ceo_info.db' did not exist, so it was created without any relations (tables).

```python
db = sqlalchemy.create_engine('sqlite:///tech_ceo_info.db')
```

In the following cell, I use pandas to create a dataframe representing our the performance of various CEOs in our degree programs data (fictitious). 

```python
dep_data = pd.DataFrame({
    "name": ["Sergey Brin", "Danah Boyd", "Bill Gates", "Hillary Mason", "Mike Olson", "Mark Zuckerburg", "Sheryl Sandberg", "Susan Wojcicki", "Marissa Mayer"],
    "gpa": [2.8, 3.9, 1, 4, 3.7, 3.8, 3.6, 3.8, 2.6],
    "age": [45, 40, 63, 39, 53, 34, 49, 50, 43],
    "dept": ["CS", "CS", "CS", "DATASCI", "CS", "CS", "BUSINESS", "BUSINESS", "BUSINESS"]
})
```

```python
dep_data
```

Now I create another Dataframe of the net worths for each tech giant (not ficticious).

```python
net_worth = pd.DataFrame({
    "name": ["Sergey Brin", "Danah Boyd", "Bill Gates", "Hillary Mason", "Mike Olson", "Mark Zuckerburg", "Sheryl Sandberg", "Susan Wojcicki", "Marissa Mayer"],
    "net_worth_millions": [82200, 6, 1100, None, 61, 48400, 15000, 765, 600]
})
```

```python
net_worth
```

Now write both of our tables into the database.

```python
dep_data.to_sql('dep_gpa_data', db)
```

```python
net_worth.to_sql('tech_ceo_net_worth', db)
```
