```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
## Load data
states = pd.read_csv("state_demographics.csv")
west_coast = states[states["State"].isin(["California", "Oregon", "Washington"])]
iris = pd.read_table('iris.txt', sep = ",")
iris.columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
tips = sns.load_dataset("tips")
dj = sns.load_dataset("dowjones")
```

#### Visualizing distributions

```python
sns.histplot(states["Income.Median Household Income"], bins = 9)
plt.xlabel("Median Household Income")
plt.title("Median Income in U.S. States");
```

```python
sns.kdeplot(states["Income.Median Household Income"])
plt.xlabel("Median Household Income")
plt.axvline(np.median(states["Income.Median Household Income"]), linestyle = "dashed", color = "black")
plt.title("Median Income in U.S. States");
```

```python
sns.histplot(states["Income.Median Household Income"], bins = 9, stat = "density")
plt.xlabel("Median Household Income")
plt.title("Median Income in U.S. States");
```

Encode a categorical variable with color to plot different distributions for each level of the variable

```python
sns.histplot(data = iris, x="sepal_length", hue="species", bins = 20, alpha = 0.5)
plt.xlabel("Sepal Length (mm)")
plt.title("Sepal Length for Iris species");
```

By calling multiple a figure multiple times, can create overlaying data

```python
sns.kdeplot(data=iris[iris["species"] == "Iris-setosa"], x="sepal_length", fill = True)
sns.kdeplot(data=iris[iris["species"] == "Iris-versicolor"], x="sepal_length", fill = True)
sns.kdeplot(data=iris[iris["species"] == "Iris-virginica"], x="sepal_length", fill = True)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (10,4))
sns.kdeplot(data=iris[iris["species"] == "Iris-setosa"], x="sepal_length", fill = True, ax = axes[0])
sns.kdeplot(data=iris[iris["species"] == "Iris-versicolor"], x="sepal_length", fill = True, ax = axes[1])
sns.kdeplot(data=iris[iris["species"] == "Iris-virginica"], x="sepal_length", fill = True, ax = axes[2])
```

#### Two quantitative variables

```python
with sns.axes_style("whitegrid"):
    sns.relplot(data = states, x = "Education.Bachelor's Degree or Higher", y = "Income.Median Household Income")
    plt.xlabel("Percentage of Population with Bach. Deg. or Higher")
    plt.ylabel("Median Household Income");
```

```python
with sns.axes_style("whitegrid"):
    sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker")
    plt.ylabel("Tip Amount")
    plt.xlabel("Total Bill");
```

```python
sns.relplot(data = dj, x = "Date", y = "Price")
plt.xticks(ticks = ["1914-12-01", "1924-12-01", "1934-12-01", "1944-12-01", "1954-12-01", "1964-12-01"],
          labels = ["1914","1924","1934","1944","1954","1964"])
plt.ylabel("Dow Jones Index");
```

```python
sns.lineplot(data = dj, x = "Date", y = "Price")
plt.xticks(ticks = ["1914-12-01", "1924-12-01", "1934-12-01", "1944-12-01", "1954-12-01", "1964-12-01"],
          labels = ["1914","1924","1934","1944","1954","1964"])
plt.ylabel("Dow Jones Index");
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(iris.sepal_width, iris.sepal_length, 
            c = iris.petal_length, s=(iris.petal_width**2)*60, cmap="viridis")
ax = plt.gca()

plt.colorbar(label="Petal Length")
plt.xlabel("Sepal Width")
plt.ylabel("Sepal Length")

#make a legend:
pws = [0.5, 1, 1.5, 2., 2.5]
for pw in pws:
    plt.scatter([], [], s=(pw**2)*60, c="k",label=str(pw))

h, l = plt.gca().get_legend_handles_labels()
plt.legend(h[1:], l[1:], labelspacing=1.2, title="Petal Width", borderpad=1, 
            frameon=True, framealpha=0.6, edgecolor="k", facecolor="w")

plt.show()
```
