```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

Just plotting the data will use seaborn's default color palette (which is decent!).

```python
sns.color_palette()
```

The colors are stored as RGB values, scaled to [0,1]

```python
sns.color_palette()[0]
```

```python
penguins = penguins[~penguins["sex"].isna()]
sns.swarmplot(data = penguins, x = "body_mass_g", y="sex", hue="species");
```

We can specifiy our own colors using names that seaborn recognizes. These are common across many plotting frameworks. 

```python
palette = ["forestgreen","violet","coral"]
```

```python
sns.color_palette(palette)
```

```python
sns.swarmplot(data = penguins, x = "body_mass_g", y="sex", hue="species",
           palette= sns.color_palette(palette, 3));
```

```python
sns.color_palette("tab10")
```

We can also use hex colors for more control over our colors. 

![title](./images/hex_color.jpg)

Hex colors are just hexadecimal representations of RGB values.

```python
palette = ["#754cf4","#b20fe9","#e9450f"]
```

```python
sns.color_palette(palette)
```

```python
sns.stripplot(data = penguins, x = "body_mass_g", y= "bill_length_mm", hue="species",
           palette= sns.color_palette(palette, 3));
```

Alternatively, use seaborn's built in color palettes.


```python
sns.boxplot(data = penguins, x = "body_mass_g", y="sex", hue="species",
           palette= "viridis");
```

```python
sns.color_palette("pastel")
```

```python
sns.color_palette("Set2")
```

```python
sns.color_palette("viridis", as_cmap=True)
```

```python
sns.color_palette("dark:#754cf4", as_cmap=True)
```

```python
sns.color_palette("blend:#754cf4,#eb3434", as_cmap=True)
```
