```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn import datasets
```

## What is 'matplotlib'?

'matplotlib' is the most widely used data visualization library for Python. It offers a high level of control and customization over your visuals at the cost of some subpar defaults; a common tradeoff among data visualization libraries. It is object-oriented, effectively meaning that we begin by creating empty objects (plots) referred to as 'figures' or 'axes', and then tailor them with bespoke methods.

Let's look at a simple example of a scatterplot in 'matplotlib' using the 'iris' dataset: 

```python
# Load in iris data
iris = pd.read_table('./data/iris.txt', sep = ",")
iris.columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot observations from the first and second columns of the iris dataset as points on a scatterplot
ax.scatter(x = iris["sepal_length"], y = iris["sepal_width"], c = "darkblue")

# Create plot labels
ax.set_xlabel("Sepal Length (mm)")
ax.set_ylabel("Sepal Width (mm)")
ax.set_title("Iris Sepal Allometry")

# Create gridlines
ax.grid(True)

# Tighten up the figure to fill its margins 
fig.tight_layout()
```

Here we've created a very straightforward scatterplot with some basic but essential plotting elements (technically referred to as "Artists" in 'matplotlib' parlance). Note that while we did specify the details of some elements *e.g.* axis labels and point color, a number of other elements were handled automatically by 'matplotlib' *e.g.* tick markers and labels along the axes, grid spacing, etc...

In general, any elements left unspecified by the user will be handled by 'matplotlibs' defaults. For now we will focus on building a plot with its most basic elements. 

In 'matplotlib', figures can be created in an "element-wise" manner, meaning that we can build our figure piece-wise to understand how each "artist" fits into the bigger picture. 

First, lets simply initialize the empty figure.

```python
# Create empty figure
fig, ax = plt.subplots()
```

The default behavior when we call `plt.subplots()` is to create an empty plot (a set of 2 "Axes") with some very simple defaults for each axis. We've initialized the object and started plotting, and now we need to give it our data. We do this by calling a method on our object "ax", in this case `ax.scatter()`, indicating that we wish to visualize our observations as a scatterplot. 

Note that we pass three arguments: `x`, `y`, and `c` (for color). We could have passed fewer arguments, and there are many more we could have used to further customize our plot. Check out `help(ax.scatter)` if you're curious. 

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot observations from iris as a scatter plot
ax.scatter(x = iris["sepal_length"], y = iris["sepal_width"], c = "blue");
```

Our data is now displayed on the previously empty plot. As you can see, 'matplotlib' automatically adjusted the axis ranges and tick labels to appropriately capture the full range of our two variables.

With data drawn, lets now add some much needed context. We'll use the `set_xlabel`, `set_ylabel` and `set_title` methods to properly label our plot. 

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot observations from the first and second columns of the iris dataset as points on a scatterplot
ax.scatter(x = iris["sepal_length"], y = iris["sepal_width"], c = "blue");

# Create plot labels
ax.set_xlabel("Sepal Length (mm)")
ax.set_ylabel("Sepal Width (mm)")
ax.set_title("Iris Sepal Allometry")
```

We could stop here and feel pretty good about our visualization, but lets just add a couple finishing touches. First, lets add gridlines to the background. This is primarily aesthetic preference, but gridlines can aid the comparison of values.

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot observations from the first and second columns of the iris dataset as points on a scatterplot
ax.scatter(x = iris["sepal_length"], y = iris["sepal_width"], c = "blue");

# Create plot labels
ax.set_xlabel("Sepal Length (mm)")
ax.set_ylabel("Sepal Width (mm)")
ax.set_title("Iris Sepal Allometry")

# Create gridlines
ax.grid(True)
```

Finally, lets make one final, minor adjustment. `tight_layout()` automatically adjusts padding in the plot margins to create more room for axis labels, which aren't automically considered when the plot is first initialized. This function becomes particularly relevant when plotting multiples subplots in the same figure, but also tends to adjust things nicely for single plots as well.  

```python
# Create empty figure
fig, ax = plt.subplots()

# Plot observations from the first and second columns of the iris dataset as points on a scatterplot
ax.scatter(x = iris["sepal_length"], y = iris["sepal_width"], c = "blue");

# Create plot labels
ax.set_xlabel("Sepal Length (mm)")
ax.set_ylabel("Sepal Width (mm)")
ax.set_title("Iris Sepal Allometry")

# Create gridlines
ax.grid(True)

# Tighten up the figure to fill its margins 
fig.tight_layout()
```

This is example shows we can work up a full plot using the individual building-blocks of matplotlib. This can all be tedious and unnecessary for simple plots however. This is why we often turn to frameworks like "seaborn" which provide us some sensible and attractive defaults. 

```python
sns.relplot(data = iris, x = "sepal_length", y = "sepal_width");
```

Seaborn is built on top of matplotlib, so we can further customize our plots using the same building blocks

```python
sns.relplot(data = iris, x = "sepal_length", y = "sepal_width")
plt.grid()
plt.ylabel("Sepal Length (mm)")
plt.xlabel("Sepal Width (mm)");
```

Seaborn also lets us automatically encode values based on other grouping variables in our data.

```python
sns.relplot(data = iris, x = "sepal_length", y = "sepal_width", hue = "species")
plt.grid()
plt.ylabel("Sepal Length (mm)")
plt.xlabel("Sepal Width (mm)");
```
